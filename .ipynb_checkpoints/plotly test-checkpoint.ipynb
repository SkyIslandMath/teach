```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.figure_factory as ff
```

```python
def row2fips(row):
    if row['County']<10:
        return f"0{row['State Code']}00{row['County']}"
    else:
        return f"0{row['State Code']}0{row['County']}"
```

```python
usgeo=pd.read_excel("D:/data/all-geocodes-v2018_clean.xlsx")
```

```python
usgeo.head()
```

```python
az=usgeo[usgeo['State Code']==4].reset_index()
```

```python
fips=[]
for idx,row in az.iterrows():
    if row['County']!=0:
        fips.append(row2fips(row))
fips
```

```python
ff.create_choropleth(fips=fips,values=range(len(fips)),scope=['AZ'])
```
