```python
species=df.groupby('Tree Species').mean()
```

```python
species=species.reset_index()
```

```python
species=species.drop('Tree ID',axis=1)
```

```python

species
```

```python
sp_melt=species.melt(id_vars=['Tree Species'],value_vars=['4/29/2016','10/20/2016','4/10/2017','9/11/2017'],var_name='Date',value_name='Height(in.)')
sp_melt
```

```python
sp_melt['Date']=pd.to_datetime(sp_melt['Date'])
```

```python
stdate=sp_melt['Date'].iloc[0]
```

```python
sp_melt['Days']=(sp_melt['Date']-stdate).dt.days
```

```python
sp_list=np.unique(sp_melt['Tree Species'])
sp_list=[sp for sp in sp_list if sp!='Screwbean Mesquite/Ironwood']
sp_list
#plt_list=[]
```

def get_data(species):
    mask=sp_melt['Tree Species']==species
    x=sp_melt[mask]['Days'].values
    y=sp_melt[mask]['Height(in.)'].values
    return (x,y)

data={species:get_data(species) for species in sp_list}
data.pop('Screwbean Mesquite/Ironwood')

regs={species:LinearRegression().fit(data[species][0].reshape(1,-1)
                                     ,data[species][1].reshape(1,-1)) for species in sp_list}

```python
regs=LinearRegression()
```

```python
X=sp_melt['Days']
y=sp_melt['Height(in.)']
regs.fit(X,y)
```

```python
def update(attr,old,new):
    mask=sp_melt['Tree Species']==new
    fig.renderers[0].data_source.data['x']=sp_melt[mask]["Days"].values
    fig.renderers[0].data_source.data['y']=sp_melt[mask]["Height(in.)"].values
    push_notebook()
```

```python
#trees=widgets.Dropdown(options=sp_list,description='Tree Species',value='Blue Palo Verde')
fig=sp_melt[sp_melt['Tree Species']==sp_list[0]].plot_bokeh.scatter(x='Days',y='Height(in.)',category='Tree Species',ylim=(0,200),show_figure=False)
tree=Select(title='Tree Species',value='Blue Palo Verde',options=sp_list)
tree.on_change('value',update)
l=layout([fig,tree])

#trees.observe(on_change,'value')
#display(trees)
```

```python
show(l,notebook_handle=True)
```
